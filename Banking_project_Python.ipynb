Numerical Analysis

```python
numerical_cols= ['Estimated_Income','Superannuation_Savings']

#Univariate analysis and visualization

plt.figure(figsize=(15,10)) 
for i,col in enumerate(numerical_cols):
    plt.subplot(1,2,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()      
```

```python
numerical_cols= ['Estimated_Income','Superannuation_Savings','Credit_Card_Balance','Bank_Loans','Bank_Deposits','Checking_Accounts','Saving_Accounts','Foreign_Currency_Account','Business_Lending']

#Univariate analysis and visualization

plt.figure(figsize=(15,10)) 
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()    
```

Heatmaps

```python
numerical_cols= ['Estimated_Income','Superannuation_Savings','Credit_Card_Balance','Bank_Loans','Bank_Deposits','Checking_Accounts','Saving_Accounts','Foreign_Currency_Account','Business_Lending']
correlation_matix=df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matix,annot=True,cmap='crest',fmt=".2f")
plt.title("correlation_matrix")
plt.show()

```
